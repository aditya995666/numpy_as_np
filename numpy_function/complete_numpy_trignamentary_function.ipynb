```python
x_sin=np.arange(0,3*3,0.1)
x_sin
```

```python
y_sin=np.sin(x_sin)
y_sin
```

```python
plt.plot(x_sin,y_sin)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x_cos=np.arange(0,56,0.1)
x_cos
```

```python
y_cos=np.cos(x_cos)
y_cos
```

```python
plt.plot(x_cos,y_cos)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```
