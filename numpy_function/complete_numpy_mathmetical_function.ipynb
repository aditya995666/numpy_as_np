```python
import numpy as np
```

MATHMETICAL FUNCTION

```python
arr_1 = np.arange(1,10).reshape(3,3)
arr_1
```

```python
arr_2 = np.arange(5,14).reshape(3,3)
arr_2
```

```python
add_value=np.add(arr_1,arr_2)
add_value
```

```python
sub_val=np.subtract(arr_1,arr_2)
sub_val
```

```python
mul_val=np.multiply(arr_1,arr_2)
mul_val
```

```python
index_max=np.argmax(arr_1)
index_max
```

```python
# product of matrix
arr_1 @ arr_2
```

```python
# find maximum value
index_min=np.argmin(arr_1)
index_min
```

```python
sum_=np.sum(arr_1)
sum_
```

```python
# sum all value in the column
sum_1=np.sum(arr_1,axis=0)
sum_1
```

```python
# mean value
mean=np.mean(arr_1)
mean
```

```python
squart=np.sqrt(arr_1)
squart
```

```python
exp=np.exp(arr_1)
exp
```

```python
log=np.log(arr_1)
arr_1
```

```python
log_10=np.log10(arr_1)
```
